```python
from random import randrange

print("""Welcome to The mastermind game.I pick four colors from these colors:
(r)ed, (b)lue, (y)ellow, (g)reen, (p)urple, (m)agenta 
You'll guess four colors what I picked. You have 13 chance. Let's begin!!!""")
print("""\nNOTICE: w and b variables indicate whether your guess is right, or not
For example, my colors are "rbyg", and your guess is "rmpy"
In this state: b:1 because r is right color, and first place.
               w:1 because y is right color, but the place is not true""")

colors=["r","b","y","g","p","m"]
pickedColors=[]
while len(pickedColors) < 4:
  i=colors[randrange(0,5)]
  #print(i)
  if pickedColors.count(i) < 2:
    pickedColors.append(i)
print("\nHint: My colors = " , pickedColors)

def checkinList(userGuess):
  """Check guess in picked colors"""
  t=0
  for i in userGuess:
    if i in colors:
      t += 1
  if t==4:
    return True
  else:
    return False
  
def checkinSameChar(userGuess):
  """Check guess there are same characters more than two"""
  for i in userGuess:
    if userGuess.count(i)<=2:
      return True
    else:
      return False

def checkinPlaces(userGuess):
  """Check guess is right places or not"""
  b=0
  for i in range(4):
    if userGuess[i]==pickedColors[i]:
      b+=1
  return b

def checkinColors(userGuess):
  """Check guess is right colors or not"""
  w=0
  for i in userGuess:
    if i in pickedColors:
      if userGuess.count(i)<=pickedColors.count(i):
        w+=1
  return w

def checkinAnswer(userGuess):
  """Check guess is right or not"""
  w=checkinColors(userGuess)
  b=checkinPlaces(userGuess)
  if b==4:
    return True
  else:
    print("b: {}, w: {}".format(b,(w-b)))

turnCount=1
while True:
  guess=input("Turn {}: ".format(turnCount))
  guess=guess.lower()
  if guess.isalpha():
    if len(guess) == 4:
      if checkinSameChar(guess):
        if checkinList(guess): #hataları kullanıcıya ayrı aktarmak için farklı bir yerde kontrol uyguladım.
          if checkinAnswer(guess):
            print("b: 4, w: 0 \nCongratulations!!!Your guess is correct!")
            break
          else:
            turnCount +=1 
        else:
          print("Error!Only letters (r)ed (b)lue (y)ellow (g)reen (p)urple (m)agenta allowed!")
          turnCount +=1
      else:
        print("Error!Your guess must contain no more than two same characters")
    else:
      print("Error!Your guess should be four colors. Like 'pmrb'")
      turnCount +=1
  else:
    print("Error! Only letters allowed!")
    turnCount +=1
  if turnCount == 13:
    print("You fail. Try again!")
    break
```
